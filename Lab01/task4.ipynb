```python
def sortSchedules(schedule):
  length=len(schedule)                  
  for i in range(length-1):               #using BUBBLESORT
    for k in range(length-i-1):
      if dep_time_comparision(schedule[k], schedule[k+1]) > 0:
        schedule[k], schedule[k+1]=schedule[k+1], schedule[k]
  return schedule

def dep_time_comparision(s1, s2):
  if s1[0] < s2[0]:
        return -1
  elif s1[0] > s2[0]:          
        return 1
  else:
        if s1[1] > s2[1]:
            return -1
        elif s1[1] < s2[1]:
            return 1
        else:
            return 0
input_file=open('/content/input4.txt','r')
output_file=open('output4.txt','w')

train_ft = []
testcase=int(input_file.readline().strip())
for i in range(testcase):
        line = input_file.readline().strip().split()
        name = line[0]
        departure = line[6]
        location=line[4]
        train_ft.append((name, departure, location))

sorted_schedule=sortSchedules(train_ft)
for idx in sorted_schedule:
        name = ''.join(idx[0])
        departure = idx[1]
        location=idx[2]
        print(f"{name} will departure at {location} {departure}", file=output_file)
output_file.close()


```
