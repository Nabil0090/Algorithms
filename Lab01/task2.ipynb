```python
def bubbleSort(arr):

  len_arr=len(arr)

  for i in range(len_arr-1):
    flag=0
    for i in range(len_arr-i-1):
      if arr[i]>arr[i+1]:
        arr[i],arr[i+1]=arr[i+1],arr[i]
        flag=1
    if flag==0:
      break
  return arr
  
input_file=open('/content/input2.txt','r')
output_file=open('output2.txt','w')
lines=input_file.readlines()
n=int(lines[0])
new_arr=list(map(int, lines[1].split()))
sorted_arr=bubbleSort(new_arr)
for i in range(n):
  print(sorted_arr[i], end=' ', file=output_file)
output_file.close()
```
