```python
input_file=open('/content/input1b.txt','r')
output_file=open('output1b.txt','w')
lines=int(input_file.readline())

for i in range(1,lines+1):
    expression=input_file.readline()
    expression=expression.split(' ')
    operator=expression[2]
    operand1=int(expression[1])
    operand2=int(expression[3])

    if operator == '+':
        result = operand1 + operand2
        print(f"The result of {operand1} {operator} {operand2} is {result}", file=output_file)
    elif operator == '-':
        result = operand1 - operand2
        print(f"The result of {operand1} {operator} {operand2} is {result}", file=output_file)
    elif operator == '*':
        result = operand1 * operand2
        print(f"The result of {operand1} {operator} {operand2} is {result}", file=output_file)
    elif operator == '/':
        result = operand1 / operand2
        print(f"The result of {operand1} {operator} {operand2} is {result}", file=output_file)

output_file.close()




```
