```python
def selectionSort(ids, marks):
    length = len(marks)
    
    for i in range(length):
        max_idx = i
        for j in range(i+1, length):
            if marks[j]>marks[max_idx] or (marks[j]==marks[max_idx] and ids[j]<ids[max_idx]):
                max_idx = j
        
        ids[i], ids[max_idx] = ids[max_idx], ids[i]
        marks[i], marks[max_idx] = marks[max_idx], marks[i]
    
    return ids, marks

input_file=open('/content/input3.txt','r')
output_file=open('output3.txt','w')
lines=input_file.readlines()
n=int(lines[0])
unsorted_ids=list(map(int, lines[1].split()))
unsorted_marks=list(map(int, lines[2].split()))
sorted_ids, sorted_marks=selectionSort(unsorted_ids, unsorted_marks)
for i in range(n):
    print(f"ID: {sorted_ids[i]} Mark: {sorted_marks[i]}", file=output_file)
output_file.close()
```
