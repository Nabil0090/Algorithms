```python
#task3
import heapq

def dijkstra(graph, source, destination):
    distances = [float('inf')] * len(graph)
    distances[source] = 0
    min_heap = [(0, source)]

    while min_heap:
        dist, node = heapq.heappop(min_heap)



        for neighbor, weight in graph[node]:
            danger_level = max(distances[node], weight)
            if danger_level < distances[neighbor]:
                distances[neighbor] = danger_level
                heapq.heappush(min_heap, (danger_level, neighbor))

    return distances[destination]
input3 = open("input3.txt","r")
output3 = open("output3.txt","w")
temp = [int(i) for i in input3.readline().split(" ")]
n = int(temp[0])
m = int(temp[1])

graph = [[] for _ in range(n+1)]


for _ in range(m):
        testcase=input3.readline()
        u, v, w = map(int, testcase.split())
        graph[int(u)].append((int(v), int(w)))

min_danger_level = dijkstra(graph, 1, n)
if min_danger_level == float('inf'):
        print("Impossible", file=output3)
else:
        print(min_danger_level, file=output3)









```
