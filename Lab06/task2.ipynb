```python
#task2

import heapq

def dijkstra(graph,node):
    distances=[float('inf')]*(n+1)
    distances[node]=0

    queue=[(0,node)]

    while queue:
        current_distance,current_node=heapq.heappop(queue)
        # relaxation
        for next_node,weight in graph[current_node]:
            distance_temp=current_distance+weight
            if distance_temp<distances[next_node]:
                distances[next_node]=distance_temp

                heapq.heappush(queue,(distance_temp,next_node))
    return distances


input2 = open("input2.txt","r")
output2 = open("output2.txt","w")
temp = [int(i) for i in input2.readline().split(" ")]
n = int(temp[0])
m = int(temp[1])

graph = [[] for _ in range(n+1)]


for _ in range(m):
        testcase=input2.readline()
        u, v, w = map(int, testcase.split())
        graph[int(u)].append((int(v), int(w)))

testcase=input2.readline()
testcase=list(map(int, testcase.split()))
start1=testcase[0]
start2=testcase[1]

alice_distances = dijkstra(graph, start1)
bob_distances = dijkstra(graph, start2)

min_time = float('inf')
meeting_node = -1
for node in range(1, n + 1):
    if alice_distances[node] != float('inf') and bob_distances[node] != float('inf'):
        total_time = max(alice_distances[node], bob_distances[node])
        if total_time < min_time:
            min_time = total_time
            meeting_node = node

if meeting_node == -1:
        print("Impossible", file=output2)
else:
        print("Time", min_time, file=output2)
        print("Node", meeting_node, file=output2)
```
