```python
#task2

def maximum_activities(tasks, m):
    tasks.sort(key=lambda x: x[1])
    assigned = [0] * m
    completed = 0

    for task in tasks:
        for i in range(m):
            if assigned[i] <= task[0]:
                assigned[i] = task[1]
                completed += 1
                break

    return completed

input2=open('input2.txt','r')
output2=open('output2.txt','w')

testcase=input2.readline()
n, m = map(int, testcase.split())
tasks = []
for _ in range(n):
    testcase=input2.readline()
    start_time, end_time = map(int, testcase.split())
    tasks.append((start_time, end_time))


result = maximum_activities(tasks, m)
print(result, file=output2)
```
