```python
#task1

def maximum_tasks(tasks):
    tasks.sort(key=lambda x: x[1])
    current_time = 0
    task_done = []

    for task in tasks:
        start_time, end_time = task
        if start_time >= current_time:
            task_done.append(task)
            current_time = end_time

    return task_done

input1=open('input1.txt','r')
output1=open('output1.txt','w')

N = int(input1.readline())
tasks = []
for _ in range(N):
    testcase=input1.readline()
    start_time, end_time = map(int, testcase.split())
    tasks.append((start_time, end_time))


completed_tasks = maximum_tasks(tasks)


print(len(completed_tasks), file=output1)
for task in completed_tasks:
    print(task[0], task[1], file=output1)
```
