```python
#task3-disjoint set data structure

input3 = open("input3.txt","r")
output3 = open("output3.txt","w")

temp = [int(i) for i in input3.readline().split(" ")]
n = int(temp[0])
k = int(temp[1])

parent = [i for i in range(n+1)]
size = [1] * (n+1)

def find_parent(node):
        if node == parent[node]:
            return node
        parent[node] = find_parent(parent[node])
        return parent[node]

def union_by_size(u, v):
        ulp_u = find_parent(u)
        ulp_v = find_parent(v)
        if ulp_u == ulp_v:
            return size[ulp_u]
        if size[ulp_u] < size[ulp_v]:
            parent[ulp_u] = ulp_v
            size[ulp_v] += size[ulp_u]
            return size[ulp_v]
        else:
            parent[ulp_v] = ulp_u
            size[ulp_u] += size[ulp_v]
            return size[ulp_u]

for _ in range(k):
    testcase=input3.readline()
    a, b = map(int, testcase.split())
    result = union_by_size(a, b)
    print(result, file=output3)
```
