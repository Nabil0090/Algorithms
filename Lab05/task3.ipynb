```python
#task3

input3 = open("input3.txt","r")
output3 = open("output3.txt","w")
temp = [int(i) for i in input3.readline().split(" ")]
n = int(temp[0]) #vertices
m = int(temp[1]) #edges
graph = [[] for _ in range(n+1)]
rev_graph=[[] for _ in range(n+1)]

for i in range(0,m):
  testcase = input3.readline()
  temp1=list(map(int, testcase.strip().split()))
  graph[int(temp1[0])].append(int(temp1[1]))
  rev_graph[int(temp1[1])].append(int(temp1[0]))
# print(graph)
# print(rev_graph)
def dfs1(node, graph, visited, stack):
    visited[node] = True
    for neighbor in graph[node]:
        if not visited[neighbor]:
            dfs1(neighbor, graph, visited, stack)
    stack.append(node)

def dfs2(node, graph, visited, component):
    visited[node] = True
    component.append(node)
    for neighbor in graph[node]:
        if not visited[neighbor]:
            dfs2(neighbor, graph, visited, component)
def find_strongly_connected_components(graph):

    visited_1 = [False] * (n + 1)
    stack = []
    for i in range(1, n + 1):
        if not visited_1[i]:
            dfs1(i, graph, visited_1, stack)

    visited_2 = [False] * (n + 1)
    strongly_connected_components = []
    while stack:
        node = stack.pop()
        if not visited_2[node]:
            component = []
            dfs2(node, rev_graph, visited_2, component)
            strongly_connected_components.append(component)

    return strongly_connected_components
scc = find_strongly_connected_components(graph)


for i in scc:
  print(*i, file=output3)
```
