```python
#task1b

from collections import deque
input1b = open("input1b.txt","r")
output1b = open("output1b.txt","w")
temp = [int(i) for i in input1b.readline().split(" ")]

n = int(temp[0]) #vertices
m = int(temp[1]) #edges



def bfs(n, m):
    graph = [[] for _ in range(n + 1)]
    in_degree = [0] * (n + 1)

    for i in range(0,m):
        testcase = input1b.readline()
        temp1=list(map(int, testcase.strip().split()))
        graph[int(temp1[0])].append(int(temp1[1]))
        in_degree[temp1[1]] += 1

    queue = deque()
    for i in range(1, n + 1):
        if in_degree[i] == 0:
            queue.append(i)
    order = []
    while queue:
        course = queue.popleft()
        order.append(course)
        for next_course in graph[course]:
            in_degree[next_course] -= 1
            if in_degree[next_course] == 0:
                queue.append(next_course)

    if len(order) == n:
        return order
    else:
        return None
order = bfs(n,m)

if order:
    print(*order, file=output1b)
else:
    print("IMPOSSIBLE", file=output1b)
```
