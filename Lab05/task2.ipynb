```python
def lexSeq(g, indeg, vtx):
    queue = []
    count = 0
    txt = ''

    for i in range(1, len(indeg)):
        if indeg[i] == 0:
            queue.append(i)

    while queue:
        smol = min(queue)
        idx = queue.index(smol)
        u = queue.pop(idx)
        txt += str(u) + ' '
        count += 1
        for v in g[u]:
            indeg[v] -= 1
            if indeg[v] == 0:
                queue.append(v)
    if count != vtx:
        return 'IMPOSSIBLE'
    else:
        return txt


input2 = open('input2.txt', 'r')
output2 = open('output2.txt', 'w')

vtx, edg = list(map(int, input2.readline().split()))
adj = [0]*(vtx+1)
indeg = [0]*(vtx+1)

for i in range(len(adj)):
    adj[i] = []

for i in range(edg):
    u, v = list(map(int, input2.readline().split()))
    adj[u].append(v)
    indeg[v] = indeg[v]+1

res = lexSeq(adj, indeg, vtx)
print(res, file=output2)
output2.close()
```
