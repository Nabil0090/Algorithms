```python
#task1a

input1a = open("input1a.txt","r")
output1a = open("output1a.txt","w")
temp = [int(i) for i in input1a.readline().split(" ")]
n = int(temp[0]) #vertices
m = int(temp[1]) #edges
graph = [[] for _ in range(n+1)]

for i in range(0,m):
  testcase = input1a.readline()
  temp1=list(map(int, testcase.strip().split()))
  graph[int(temp1[0])].append(int(temp1[1]))

visited=[False] * (n+1)
stack=[]

def depth_first_search(node, visited, graph):
    visited[node] = True
    for i in graph[node]:
        if not visited[i]:
            depth_first_search(i, visited, graph)

    stack.append(node)


def topological_sort(graph, visited):
    for v in range(1, n+1):
        if not visited[v]:
            depth_first_search(v, visited, graph)
def print_stack(stack):
    while stack:
        current = stack.pop()
        print(current, end=" ", file=output1a)
topological_sort(graph, visited)

print_stack(stack)
```
