```python
#task2b
def mergealgo(a,b,m,n):
  c=[0]*(m+n)
  i=0
  j=0
  k=0
  while i<m and j<n:
    if a[i]<b[j]:
      c[k]=a[i]
      k+=1
      i+=1
    else:
      c[k]=b[j]
      k+=1
      j+=1
  while i<m:
    c[k]=a[i]
    k+=1
    i+=1
  while j<n:
    c[k]=b[j]
    k+=1
    j+=1

  return c
input_file=open('/content/input2b.txt','r')
output_file=open('output2b.txt','w')

testcase=input_file.readlines()
ali_arr=list(map(int, testcase[1].strip().split()))
bob_arr=list(map(int, testcase[3].strip().split()))
ali_size=int(testcase[0])
bob_size=int(testcase[2])


merge_arr=mergealgo(ali_arr,bob_arr,ali_size,bob_size)
for i in merge_arr:
  print(i, end=" ", file=output_file)
output_file.close()
```
