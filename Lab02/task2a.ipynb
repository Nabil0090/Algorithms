```python
#task2a

input_file=open('/content/input2a.txt','r')
output_file=open('output2a.txt','w')
testcase=input_file.readlines()
ali=list(map(int, testcase[1].strip().split()))
bob=list(map(int, testcase[3].strip().split()))
merg_arr=[]
for i in ali:
  merg_arr.append(i)
for j in bob:
  merg_arr.append(j)
merg_arr.sort()                         #the python builtin sort() function runs on nlogn time. Thus total running time nlogn.
for i in range(len(merg_arr)):
  print(merg_arr[i], end=' ',file=output_file)
```
