```python
#task4
def max_num(arr,left, right):
  if left==right:
    return arr
  if left+1==right:
    if arr[left]<arr[right]:
      return arr[right]
    else:
      return arr[left]
  mid=(left+right)//2
  left_max=max_num(arr, left, mid)
  right_max=max_num(arr, mid+1, right)



  max_val = left_max if left_max > right_max else right_max

  return max_val
input_file=open('/content/input4.txt','r')
output_file=open('output4.txt','w')
testcase=input_file.readlines()
size=int(testcase[0])
array=list(map(int, testcase[1].strip().split()))

print(max_num(array,0,size-1), file=output_file)
output_file.close()

#time complexity is nlogn cz im using divide and conquer method simultaneously.
```
