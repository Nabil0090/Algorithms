```python
#task1a
input_file=open('/content/input1a.txt','r')
output_file=open('output1a.txt','w')

size, target = map(int, input_file.readline().split())
arr = list(map(int, input_file.readline().split()))


l=len(arr)
flag=0
for i in range(l):
  for j in range(i+1,l):
    if arr[i]+arr[j]==target:
      print(f"{i+1} {j+1}", file=output_file)
      flag=1
      break
if flag==0:
  print("Impossible", file=output_file)
output_file.close()

```
