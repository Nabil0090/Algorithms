```python
input1 = open("input1a.txt","r")
output1 = open("output1a.txt","w")
temp = [int(i) for i in input1.readline().split(" ")]
n = int(temp[0])
m = int(temp[1])
matrix=[[0]*(n+1) for i in range(n+1)]
for i in range(0,m):
  testcase=input1.readline()
  temp1=list(map(int, testcase.strip().split()))
  matrix[temp1[0]][temp1[1]] = temp1[2]
for row in matrix:
    print(' '.join(map(str, row)), file=output1)
output1.close()
```
