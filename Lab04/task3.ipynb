```python
#task3
input3 = open("input1b.txt","r")
output3 = open("output1b.txt","w")
temp = [int(i) for i in input3.readline().split(" ")]
n = int(temp[0]) #vertices
m = int(temp[1]) #edges
graph = [[] for _ in range(n+1)]

for i in range(0,m):
  testcase = input3.readline()
  temp1=list(map(int, testcase.strip().split()))
  graph[int(temp1[0])].append(int(temp1[1]))
  graph[int(temp1[1])].append(int(temp1[0]))
visited=[False]*n
def dfs(visited, graph, node):
  if node not in visited:
    print(node, end=" ", file=output3)
    visited[node]=True
    for nei in graph[node]:
      dfs(visited, graph, nei)
dfs(visited, graph, 1)
```
