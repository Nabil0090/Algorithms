```python
input6 = open("input6.txt",'r')
output6 = open("output6.txt",'w')
rows, columns = map(int, input6.readline().rstrip().split())
g = []
for line in input6:
    g.append([i for i in line.rstrip()])

def DFS(i, j):
    global columns, rows
    if i<0 or i>=rows or j<0 or j>=columns:
        return
    elif g[i][j] == '#':
        return
    else:
        if g[i][j] == 'D':
            global diamonds
            diamonds+=1
        g[i][j] = '#'
        DFS(i+1, j)
        DFS(i-1, j)
        DFS(i, j+1)
        DFS(i, j-1)


def floodFill():
    global rows, columns, diamonds
    diamondsArr = []
    for i in range(rows):
        for j in range(columns):
            diamonds = 0
            if g[i][j] != "#":
                DFS(i, j)
                diamondsArr.append(diamonds)
    return diamondsArr

maxDiamonds = max(floodFill())
output6.write(str(maxDiamonds))

output6.close()
```
