```python
#task1b
input1 = open("input1b.txt","r")
output1 = open("output1b.txt","w")
temp = [int(i) for i in input1.readline().split(" ")]
n = int(temp[0])
m = int(temp[1])
adjlist = [[] for _ in range(n+1)]

for i in range(0,m):
  testcase = input1.readline()
  temp1=list(map(int, testcase.strip().split()))
  adjlist[int(temp1[0])].append((int(temp1[1]), int(temp[2])))



# print(adjlist[1][0])

for i in range(0, n + 1):
    print(f"{i} :", end=" ", file=output1)
    for neighbor, wei in adjlist[i]:
        print(f"({neighbor} {wei})", end=" ", file=output1)
    print(file=output1)
input1.close()
output1.close()
```
