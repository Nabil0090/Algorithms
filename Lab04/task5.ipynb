```python
class Queue:
    def __init__(self):
        self.queue = []
    def enqueue(self, element):
        self.queue.append(element)
    def dequeue(self):
        temp = self.queue[0]
        self.queue = self.queue[1:]
        return temp
    def peek(self):
        if len(self.queue)!=0:
            return self.queue[0]
        else:
            return None

class Vertex:
    def __init__(self, value):
        self.value = value
        self.adjacent = []
        self.parent = None
        self.distance = 0
        self.color = 0


input5=open("input4.txt","r")
output5=open("output4.txt"."w")
N,M,D = map(int, input5.readline().split())
G = {i: Vertex(i) for i in range(1, N+1)}
Q = Queue()

for i in input5:
    u, v = map(int, i.rstrip().split())
    G[u].adjacent += [v]
    G[v].adjacent += [u]

s = G[1]
s.color = 1
Q.enqueue(s)
s = G[1]
s.color = 1
Q.enqueue(s)

while Q.peek():
    current = Q.dequeue()
    for v in current.adjacent:
        v = G[v]
        if v.color == 0:
            v.color = 1
            v.parent = current
            v.distance += current.distance + 1
            Q.enqueue(v)
    current.color = 2

v = G[D]
output5.write(f"Time: {v.distance}\n")

ans = [v.value]
while v.parent:
    ans.append(v.parent.value)
    v = v.parent
ans.reverse()

output5.write("Shortest Path: ")
for i in ans:
    print(i, end = " ")
    output5.write(f"{i} ")


output5.close()
```
