```python
input4=open("input4.txt","r")
output4=open("output4.txt","w")
N, M = map(int, input4.readline().split())
lines = [tuple(map(int, input4.readline().split())) for i in range(M)]

def is_cycle(adjlst, x, s, n):
    x[n] = 1
    s[n] = 1
    for i in adjlst[n]:
        if not x[i]:
            if is_cycle(adjlst, x, s, i):
                return 1
        elif s[i]:
            return 1
    s[n] = 0
    return 0

def in_map_cycle(N, lines):
    adjlst = [[] for i in range(N + 1)]
    for u, v in lines:
        adjlst[u].append(v)
    x = [0]*(N + 1)
    s = [0]*(N + 1)
    for n in range(1, N + 1):
        if not x[n]:
            if is_cycle(adjlst, x, s, n):
                return "YES"

    return "NO"

cycle = in_map_cycle(N,lines)

output4.write(cycle)
output4.close()
```

```python
class Solution:
    def detectCycle(self, v, node, rst, vis):
        vis[node] = 1
        rst[node] = 1
        for x in v[node]:
            if vis[x] == 0 and self.detectCycle(v, x, rst, vis):
                return True
            elif rst[x] == 1:
                return True
        rst[node] = 0
        return False

    def canFinish(self, numCourses, prerequisites):
        v = [[] for _ in range(numCourses)]
        vis = [0] * numCourses
        rst = [0] * numCourses
        for x in prerequisites:
            v[x[1]].append(x[0])
        for i in range(numCourses):
            if vis[i] == 0 and self.detectCycle(v, i, rst, vis):
                return False
        return True
```

```python
def has_cycle(graph):
    def dfs(node, visited, recursion_stack):
        visited[node] = True
        recursion_stack[node] = True

        for neighbor in graph[node]:
            if visited[neighbor]==False and dfs(neighbor, visited, recursion_stack):
                    return True
            elif recursion_stack[neighbor]==True:
                return True

        recursion_stack[node] = False
        return False

    num_nodes = len(graph)
    visited = [False] * num_nodes
    recursion_stack = [False] * num_nodes

    for node in range(num_nodes):
        if not visited[node]:
            if dfs(node, visited, recursion_stack)==True:
                return True

    return False

input3 = open("input1b.txt","r")
output3 = open("output1b.txt","w")
temp = [int(i) for i in input3.readline().split(" ")]
n = int(temp[0]) #vertices
m = int(temp[1]) #edges
graph = [[] for _ in range(n+1)]

for i in range(0,m):
  testcase = input3.readline()
  temp1=list(map(int, testcase.strip().split()))
  graph[int(temp1[0])].append(int(temp1[1]))
print(has_cycle(graph))
```
